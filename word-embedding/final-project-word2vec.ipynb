```python
import metapy
```

```python
model = metapy.embeddings.load_embeddings('people_emd.toml')
```

# Obama

```python
embed = model.at("obama")
```

```python
embed[0]
```

```python
embed[1]
```

```python
embed1 = model.at('barack')
```

```python
embed1[0]
```

```python
embed1[1]
```

```python
combine = embed[1]+embed1[1]
```

```python
combine
```

```python
for i in model.top_k(embed[1]):
    print(model.term(i[0]))
```

```python
for i in model.top_k(embed1[1]):
    print(model.term(i[0]))
```

```python
for i in model.top_k(combine):
    print(model.term(i[0]), i[2])    
```
